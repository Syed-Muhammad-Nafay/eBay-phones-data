### Creating a function for plotting histogram to understand distribution.

```python
def price_distribution_histogram(price, bins, r=45):
    xticks = np.linspace(unclean_df[unclean_df['price'] < price]['price'].min(), unclean_df[unclean_df['price'] < price]['price'].max(), (bins + 1))
    plt.figure(figsize=(10, 6))
    plt.hist(unclean_df[unclean_df['price'] < price]['price'], bins = bins)
    plt.xlabel('Prices')
    plt.ylabel('Frequency')
    plt.title(f'Distribution of Item under {price}$')
    plt.xticks(xticks, rotation = r)
    plt.grid()
    plt.show
```

### OUTLIER DETECTION

```python
Q1 = unclean_df['price'].quantile(0.25)
Q3 = unclean_df['price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
is_outlier = (unclean_df['price'] < lower_bound) | (unclean_df['price'] > upper_bound)
normal_value = unclean_df[~is_outlier]
outliers = unclean_df[is_outlier]
plt.scatter(normal_value['price'], np.zeros(len(normal_value)), color='blue', label='Normal', alpha=0.2, s=2)
plt.scatter(outliers['price'], np.zeros(len(outliers)), color='red', label='Outliers', alpha=1, s=3)
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```

```python
exclude_keywords = ["case", "cover", "screen protector", "charger", "charging","cable", "glass", "tempered", "film", "dock", "stand",
        "mount", "strap", "holder", "replacement", "earbuds", "stylus", "adapter", "ear"]
def percentage_match(price, lowest_price = unclean_df['price'].min(), exclude = exclude_keywords):
    x = unclean_df[(unclean_df['price'] < price) & (unclean_df['price'] > lowest_price)]
    absolute_search = False
    for i in exclude_keywords:
        absolute_search = absolute_search | x['product_name'].str.contains(i, case= False)
    total_df = x.shape[0]
    unmatched = x[absolute_search].shape[0]
    if total_df != 0:
        return (unmatched/total_df) * 100
    else:
        return 0
```

```python
def data_divider_by_percentage(x):
    df_objects = []
    per_list =[]
    sorted = unclean_df.sort_values(by='price', ascending=True)
    total_rows = unclean_df.shape[0]
    minus_val = total_rows - (total_rows%x)
    step = minus_val / x
    for i in range(0, x):
        start = int(step * i)
        end = int(step * (i+1))
        if i != x-1:
            df_objects.append(sorted.iloc[start:end])
        else:
            df_objects.append(sorted.iloc[start:end+minus_val])
    for i in df_objects:
        percent = percentage_match(i['price'].max(), i['price'].min())
        per_list.append(percent)
    return [per_list, df_objects]
```

```python
per_list, df = data_divider_by_percentage(200)
```

## Moment of concern.
The standard deviation in prices is very high across the table, especially in the Brand New condition, which suggests that we are dealing with a diverse set of phones. This means the dataset may still be useful for other kinds of analysis, but any price-based insights would be unreliable in its current form. The high variation could either mean there are many different phone models, each with different pricing, or just a few models with large price gaps between listings. In either case, the standard deviation test alone isn’t enough to draw meaningful conclusions about price. To improve the quality of our analysis, we will now try to extract and label the phone models from the product titles

## SAMSUNG

```python
df[df['Brands'] == 'samsung'].shape
```

```python
df['Models'] = None
```

```python
def phone_model_finder(brand, x):
    var = df[(df['Brands'] == brand) & (df['product_name'].str.contains(x, case=False, na=False)) & (df['Models'].isna())].shape[0]
    return var
def phone_model_assigner(brand, x):
    df.loc[(df['Brands'] == brand) & (df['product_name'].str.contains(x, case=False, na=False)) & (df['Models'].isna()), 'Models'] = x
def phone_deletor(df, brand, x):
    return df[~((df['Brands'] == brand) & (df['product_name'].str.contains(x, case=False, na=False)) & (df['Models'].isna()))]
```

```python
models = ['s22', 'a52', 's20', 'a53', 'a23', 'a50', 'a13', 'a14', 'a40', 'a41', 'a21', 'a35', 'a55', 'a33', 's9', 'a32', 'a42', 's23', 'a54', 'a51', 'a15', 'a16', 'a71', 'a22', 'xcover', 'note']
```

```python
models_present = [f'{i} : {phone_model_finder('samsung', i)}' for i in models]
models_present
```

```python
for i in models:
    phone_model_assigner('samsung', i)
```

```python
df = df[~((df['Brands'] == 'samsung') & (df['Models'].isna()))]
```

## Cleaning the location column

```python
df['first_two_words'] = df['location_of_origin'].str.extract(r'^(\w+\s+\w+)', expand=False)
```

```python
df['first_two_words'].unique()
```

```python
dif_words = ['Located in', 'from', 'Ships from']
pattern = r'\b(?:' + '|'.join(map(re.escape, dif_words)) + r')\b'
df['location_of_origin'] = df['location_of_origin'].str.replace(pattern, '', regex=True, case=False)
```

```python
df.drop(['first_two_words', 'category'], axis=1, inplace=True)
```

### Completed with cleaning of the location column lets move on to delivery column.
This will hopefully mark the last stage in dataframe cleaning and we can resume with the analysis.

```python
df.loc[(df['delivery_charges'] == 'Free International Shipping'), 'delivery_charges'] = 0
```

```python
df['shipping_cost'] = df['delivery_charges'].str.extract(r'(\d+)', expand=False)
```

```python
df[(df['shipping_cost'].isna()) & (df['delivery_charges'] != 0)]['delivery_charges'].unique()
```

```python
df.loc[df['delivery_charges'] == 'Shipping not specified'] = 0
```

```python
df['shipping_cost'] = df['shipping_cost'].fillna(0)
df['shipping_cost'] = pd.to_numeric(df['shipping_cost'])
df['delivery_charges'] = df['shipping_cost']
df.drop(columns=['shipping_cost'], inplace=True)
```
